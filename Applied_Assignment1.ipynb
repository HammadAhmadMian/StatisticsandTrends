```python
def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0
```

```python
fifa_players_data_df['Value'] = fifa_players_data_df['Value'].apply(value_to_float)
fifa_players_data_df['Wage'] = fifa_players_data_df['Wage'].apply(value_to_float)
```

```python
def plot_random_scatter(x, y):
    plt.figure(dpi=100)
    plt.scatter(x, y, color='blue')
    plt.xlabel('Overall Rating')
    plt.ylabel('Wage per week in Euros')
    plt.show()
    return
```

```python
plot_random_scatter(fifa_players_data_df.Overall, fifa_players_data_df.Wage)
```

```python
fifa_players_data_df_age30plus = fifa_players_data_df.drop(fifa_players_data_df[fifa_players_data_df.Age < 31].index)
fifa_players_data_df_age30plus
```

```python
age_counts = fifa_players_data_df_age30plus['Age'].value_counts()
age_counts
```

```python
def plot_ages(age_counts, labels= age_counts.index):
    plt.figure(dpi=100)
    plt.pie(age_counts, labels=labels, autopct='%1.1f%%')
    plt.title('Pie chart for players aged above 30.')
    plt.axis('Equal')
    plt.show()
    return
```

```python
plot_ages(age_counts, labels = age_counts.index)
```

```python
countries_top5 = ['England', 'Spain', 'Italy', 'France', 'Germany']
players_top5_countries_df = fifa_players_data_df[fifa_players_data_df['Nationality'].isin(countries_top5)]
```

```python
players_top5_countries_df
```

```python
stats_by_nationality = players_top5_countries_df.groupby('Nationality')['Overall'].describe()
stats_by_nationality
```

```python
def plot_boxplot_top5_nationalities(players_top5_countries_df):
    # Group ratings by nationality
    ratings_by_nationality = [players_top5_countries_df[players_top5_countries_df['Nationality'] == nationality]['Overall']
        for nationality in countries_top5]
    
    # Plot the boxplot for all top 5 countries in one graph
    plt.figure(dpi=100)
    plt.boxplot(ratings_by_nationality, labels=countries_top5, patch_artist=True)
    plt.title('Boxplots for Overall Rating for Top 5 Countries')
    plt.ylabel('Overall Rating')
    plt.xlabel('Nationality')
    plt.show()
```

```python
plot_boxplot_top5_nationalities(players_top5_countries_df)
```
